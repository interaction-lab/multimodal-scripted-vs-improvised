```python
import os, re, sys
import torch
import pandas as pd
import numpy as np
import evaluate
from datasets import Dataset
from torch import nn
from transformers import (
    AutoTokenizer,
    DataCollatorWithPadding,
    TrainingArguments,
    Trainer,
    EarlyStoppingCallback,
    RobertaModel,
)

# ------------------ CONFIG ------------------
EDA_PATH = "/home1/anurades/multimodal-scripted-vs-improvised/multimodal-dac/EDA_Dataset.csv"
BASE_DIR = "/home1/anurades/multimodal-scripted-vs-improvised/multimodal-dac/RESULTS"
RUN_ID = "R01"
MODE = "text_roberta"
MODEL_NAME = "/home1/anurades/hf_models/roberta-base-full"

SAVE_DIR = f"{BASE_DIR}/{MODE}/{RUN_ID}"
os.makedirs(f"{SAVE_DIR}/model", exist_ok=True)
os.makedirs(f"{SAVE_DIR}/logs", exist_ok=True)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# ------------------ LOAD DATA ------------------
df = pd.read_csv(EDA_PATH)
df = df.dropna(subset=["utterance", "EDA", "filename"])

# ------------------ SCENARIO EXTRACTION ------------------
df_scripted = df[df["filename"].str.contains("script", case=False)].copy()

def extract_scenario_number(fname):
    m = re.search(r'script0*([0-9]+)_(\d+)', fname, re.IGNORECASE)
    if not m:
        return None
    group = int(m.group(1))
    within = int(m.group(2))
    mapping = {
        (1, 1): 1, (1, 2): 2, (1, 3): 3,
        (2, 1): 4, (2, 2): 5,
        (3, 1): 6, (3, 2): 7,
    }
    return mapping.get((group, within))

df_scripted["scenario_number"] = df_scripted["filename"].apply(extract_scenario_number)
df_scripted = df_scripted.dropna(subset=["scenario_number"])
print("Unique scenario numbers:", sorted(df_scripted["scenario_number"].unique()))

# ------------------ DEFINE SPLIT ------------------
R01_SPLIT = {"train": [1, 2, 3, 4, 5], "val": [6], "test": [7]}
train_df = df_scripted[df_scripted["scenario_number"].isin(R01_SPLIT["train"])]
val_df   = df_scripted[df_scripted["scenario_number"].isin(R01_SPLIT["val"])]
test_df  = df_scripted[df_scripted["scenario_number"].isin(R01_SPLIT["test"])]
print(f"Train: {len(train_df)}, Val: {len(val_df)}, Test: {len(test_df)}")

# ------------------ TOKENIZATION + LABELS ------------------
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
def tokenize(batch):
    return tokenizer(batch["utterance"], truncation=True, padding="max_length", max_length=128)

label2id = {label: i for i, label in enumerate(sorted(df["EDA"].unique()))}
id2label = {i: label for label, i in label2id.items()}

# compute class weights
class_counts = train_df["EDA"].value_counts()
num_labels = len(label2id)
weights = torch.zeros(num_labels, dtype=torch.float)
for label, idx in label2id.items():
    weights[idx] = 1.0 / class_counts.get(label, 1)
weights = weights / weights.sum() * num_labels
weights = weights.to(device)
print("Class weights:", weights)

def encode_labels(batch):
    batch["labels"] = label2id[batch["EDA"]]
    return batch

def prepare_dataset(df_part):
    dataset = Dataset.from_pandas(df_part[["utterance", "EDA"]])
    dataset = dataset.map(tokenize, batched=True)
    dataset = dataset.map(encode_labels)
    dataset = dataset.remove_columns(["utterance", "EDA", "__index_level_0__"])
    return dataset

train_dataset = prepare_dataset(train_df)
val_dataset   = prepare_dataset(val_df)
test_dataset  = prepare_dataset(test_df)
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

# ------------------ MODEL WITH 2-LAYER HEAD ------------------
class RobertaWithHead(nn.Module):
    def __init__(self, model_name, num_labels):
        super().__init__()
        self.roberta = RobertaModel.from_pretrained(model_name)
        self.dropout = nn.Dropout(0.3)
        self.classifier = nn.Sequential(
            nn.Linear(self.roberta.config.hidden_size, 256),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(256, num_labels)
        )

    def forward(self, input_ids=None, attention_mask=None, labels=None):
        outputs = self.roberta(input_ids=input_ids, attention_mask=attention_mask)
        pooled_output = outputs.last_hidden_state[:, 0, :]  # [CLS]
        x = self.dropout(pooled_output)
        logits = self.classifier(x)
        loss = None
        if labels is not None:
            loss_fct = nn.CrossEntropyLoss(weight=weights)
            loss = loss_fct(logits, labels)
        return {"loss": loss, "logits": logits}

model = RobertaWithHead(MODEL_NAME, num_labels=len(label2id)).to(device)

# ------------------ TRAINER + METRICS ------------------
accuracy = evaluate.load("accuracy")
f1 = evaluate.load("f1")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = logits.argmax(axis=-1)
    return {
        "accuracy": accuracy.compute(predictions=preds, references=labels)["accuracy"],
        "f1_macro": f1.compute(predictions=preds, references=labels, average="macro")["f1"],
    }

# Training config
args = TrainingArguments(
    output_dir=f"{SAVE_DIR}/model",
    eval_strategy="epoch",
    save_strategy="epoch",
    num_train_epochs=8,                 # allow up to 6, early stopping will cut off
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    learning_rate=1e-5,
    lr_scheduler_type="linear",
    warmup_ratio=0.1,
    save_total_limit=2,
    load_best_model_at_end=True,

    # Logging
    logging_dir=f"{SAVE_DIR}/logs",
    logging_strategy="epoch",
    log_level="info",
    report_to="none",
    disable_tqdm=False,
    fp16=torch.cuda.is_available(),
)

trainer = Trainer(
    model=model,
    args=args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=2)],
)

# ------------------ TRAIN + EVAL ------------------
sys.stdout.flush()
trainer.train()
metrics = trainer.evaluate(test_dataset)
pd.Series(metrics).to_json(os.path.join(SAVE_DIR, "metrics.json"))
print(metrics)


# ------------------ TRAIN + EVAL ------------------
sys.stdout.flush()
trainer.train()
# Evaluate on test set
print("\n=== FINAL TEST EVALUATION ===")
metrics = trainer.evaluate(test_dataset)
print(f"Test Accuracy: {metrics['eval_accuracy']:.4f}")
print(f"Test F1 (Macro): {metrics['eval_f1_macro']:.4f}")
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import seaborn as sns
import matplotlib.pyplot as plt

# ------------------ CONFUSION MATRIX ------------------
print("\nGenerating confusion matrix...")

# Get raw predictions on test set
predictions = trainer.predict(test_dataset)
preds = np.argmax(predictions.predictions, axis=-1)
labels = predictions.label_ids

# Compute matrix
unique_label_ids = sorted(set(labels) | set(preds))
used_labels = [id2label[i] for i in unique_label_ids]
cm = confusion_matrix(labels, preds, labels=unique_label_ids, normalize='true')
cm_df = pd.DataFrame(cm, index=used_labels, columns=used_labels)

cm_path = os.path.join(SAVE_DIR, "confusion_matrix.csv")
cm_df.to_csv(cm_path)
print(f"Confusion matrix saved to: {cm_path}")

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(cm_df, annot=True, fmt=".2f", cmap="Blues", cbar=False)
plt.title("Normalized Confusion Matrix (RoBERTa DAC)")
plt.ylabel("True label")
plt.xlabel("Predicted label")
pd.Series(metrics).to_json(os.path.join(SAVE_DIR, "metrics.json"))
print(metrics)
```
